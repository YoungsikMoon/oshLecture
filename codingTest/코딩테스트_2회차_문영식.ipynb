## 1. 하샤드 수

image.png

```python
import time

start = time.time()

def solution(x):
# 답안 작성 부분 ===============
    str_int = str(x)
    answer = 0
    for i in str_int :
        answer += int(i)
        if x%answer ==0:
            answer = True
        else :
            answer = False
# ==============================
    return answer


end = time.time()

print(solution(10)) #True
print(solution(12)) #True
print(solution(11)) #False
print(solution(13)) #False


print("작동시간:", f"{end - start:.5f} sec")
```

## 2. 음양 더하기

image.png

```python
import time

start = time.time()

def solution(absolutes, signs):
# 답안 작성 부분 ===============
    answer = 0

    for i in range(len(absolutes)):
        if signs[i] == True :
            answer += absolutes[i]
        elif signs[i] == False:
            answer += -1 * absolutes[i] 

# ==============================
    return answer

print(solution([4,7,12],[True,False,True]))
print(solution([1,2,3],[False,False,True]))

end = time.time()

print("작동시간:", f"{end - start:.5f} sec")
```

## 3. 부족한 금액 계산하기

image.png


```python
import time

start = time.time()


def solution(price, money, count):
# 답안 작성 부분 ===============
    answer = 0
    for i in range(1,count+1):
        answer += price*i
    answer = money - answer
    
    if answer < 0 :
        answer = abs(answer)
    elif answer >= 0:
        answer = answer
# ==============================
    return answer

print(solution(3,20,4)) # 10

end = time.time()

print("작동시간:", f"{end - start:.5f} sec")
```

## 4. 수박수박수박수박수박수?

image.png

```python
import time

start = time.time()

def solution(n):
# 답안 작성 부분 ===============
    word = '수박'
    answer = word * n
    answer = answer[0:n]
# ==============================
    return answer

print(solution(3))
print(solution(4))

end = time.time()

print("작동시간:", f"{end - start:.5f} sec")
```

## 5. 내적

image.png

```python
import time

start = time.time()


def solution(a, b):
# 답안 작성 부분 ===============
    answer = []
    for i in range(len(a)):
        answer.append(a[i] * b[i])
    answer = sum(answer)
# ==============================
    return answer

print(solution([1,2,3,4], [-3,-1,0,2]))
print(solution([-1,0,1], [1,0,-1]))

end = time.time()

print("작동시간:", f"{end - start:.5f} sec")
```
