# 문제 1. 숫자 비교하기


image.png

```python
import time

start = time.time()

def solution(num1,	num2):
# 답안 작성 부분 ===============
    if num1 == num2:
        answer = 1
    else :
        answer = -1
# ==============================
    return answer

end = time.time()

print(solution(2,3))
print(solution(11, 11))
print(solution(7,99))

print("작동시간:", f"{end - start:.5f} sec")
```

# 문제 2. 각도기

image.png

```python
import time

start = time.time()

def solution(angle):
# 답안 작성 부분 ===============
    if 0 < angle < 90 :
        answer = 1
    elif angle == 90:
        answer = 2
    elif 90 < angle < 180:
        answer = 3
    elif angle == 180 :
        answer = 4
    else :
        answer = "올바른 조건의 Angle을 입력하세요."
# ==============================
    return answer

end = time.time()

print(solution(70))
print(solution(91))
print(solution(180))

print("작동시간:", f"{end - start:.5f} sec")
```

# 문제 3. 짝수의 합


image.png

```python
import time

start = time.time()

def solution(n):
# 답안 작성 부분 ===============
    answer = []
    if n % 2 == 0 and ((n > 1 or n > 0)):
        answer.append(n)
        while n > 0 :
           n = n-2
           answer.append(n)
    if n%2 != 0 and (n > 1 or n > 0) :
        n = n-1
        answer.append(int(n))
        while n > 0 :
           n = n-2
           answer.append(n)
# ==============================
    return sum(answer)

end = time.time()

print(solution(10))
print(solution(4))

print("작동시간:", f"{end - start:.5f} sec")
```

# 문제 4. 머쓱이보다 키 큰 사람


image.png

```python
import time

start = time.time()

def solution(array,	height):
# 답안 작성 부분 ===============
    answer = 0
    for i in array:
        if i > height:
            answer += 1
# ==============================
    return answer

end = time.time()

print(solution([149, 180, 192, 170],	167))
print(solution([180, 120, 140],	190))

print("작동시간:", f"{end - start:.5f} sec")
```

# 문제 5. 옷가게 할인 받기

image.png

```python
import time

start = time.time()

def solution(price):
# 답안 작성 부분 ===============
    answer = 0
    if 100000 <= price < 300000:
        answer = price * 0.95
    elif 300000 <= price < 500000:
        answer = price * 0.9
    elif 500000 < price :
        answer = price * 0.8
    else : 
        answer = price
# ==============================
    return int(answer)

end = time.time()

print(solution(150000))
print(solution(580000))

print("작동시간:", f"{end - start:.5f} sec")
```
